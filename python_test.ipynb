```python
import requests
from textractcaller import call_textract
from textractcaller.t_constants import TextractAPI, TextractFeatures

# URL of the image containing a table
image_url = "https://web.stanford.edu/class/cs101/table.png"

# Download the image
response = requests.get(image_url)
if response.status_code != 200:
    raise Exception(f"Failed to download image from {image_url}")
image_bytes = response.content

# Call Amazon Textract to analyze the document for tables
textract_response = call_textract(
    file_source=image_bytes,
    api=TextractAPI.ANALYZE_DOCUMENT,
    features=[TextractFeatures.TABLES]
)

# Extract and print the table blocks from the response
tables = [
    block for block in textract_response.get("Blocks", [])
    if block.get("BlockType") == "TABLE"
]

print(f"Extracted {len(tables)} table(s) from the image:")

for idx, table in enumerate(tables, start=1):
    print(f"\nTable {idx}:")
    print(table)
```
